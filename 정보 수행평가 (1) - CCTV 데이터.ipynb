```python
import matplotlib.pyplot as plt
import csv

lists =['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']

f = open('전국cctv표준데이터.csv')
data = csv.reader(f)

y=[]

a=0
temp=[]
for line in data:
    temp.append(line)
f.close()
for i in lists:
    for row in temp:
        if i in row[1]:
            a+=int(row[4])
    y.append(a)
    a=0 
    
plt.rc('font',family='Malgun Gothic')
plt.title('서울특별시 구별 CCTV 수')
plt.figure(figsize = (25,30), dpi = 300)

plt.bar(list(range(0,len(lists))),y,color='red')
plt.xticks(list(range(0,25)),['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구','강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구'])
plt.yticks(y)

plt.savefig('data.png')

plt.show()
```
