```python
import csv
import matplotlib.pyplot as plt
f = open('201804_201804__.csv')

data = csv.reader(f)
next(data)
b = []
people = []

for row in data :
    b.append(row[0].split(' ')[1])
    people.append(row[1])
    
import matplotlib.pyplot as plt
plt.figure(figsize = (25,30), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.bar(b, people, label='  ', color = 'yellow')
plt.xlabel('')
plt.ylabel(' ')

plt.savefig('data.png')
plt.show()
```
