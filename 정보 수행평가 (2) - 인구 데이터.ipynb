```python
import csv
import matplotlib.pyplot as plt
f = open('201804_201804_주민등록인구및세대현황_월간.csv')

data = csv.reader(f)
next(data)
b = []
people = []

for row in data :
    b.append(row[0].split(' ')[1])
    people.append(row[1])
    
import matplotlib.pyplot as plt
plt.figure(figsize = (25,30), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.bar(b, people, label='구별 인구 수', color = 'yellow')
plt.xlabel('구')
plt.ylabel('인구 수')

plt.savefig('data.png')
plt.show()
```
