```python
import csv
import matplotlib.pyplot as plt
f = open('2010년_범죄발생지_현황 (1).csv')
data = csv.reader(f)
add = 0
y=[]
dummy1 = []
dummy2 = 0
for i in range(0,25) :
    y.append(0)
lists = ['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구','강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']
for row in data :
    for k in range(0,len(lists)) :
        if lists[k] in row[0] :
            dummy1 = y[k+1:]
            y = y[:k+1]
            dummy2 = y[k]
            y.pop()
            for i in list(range(1,len(row))) :
                d = row[i].replace(',','')
                d = d.replace('-','0')
                dummy2 += int(d)
            y.append(dummy2)
            y = y + dummy1
            dummy2 = 0
plt.figure(figsize = (25,30), dpi = 300)
plt.rc('font',family='Malgun Gothic')
plt.title('구별 범죄 발생수')
plt.xticks(range(25),lists)
plt.bar(list(range(0,len(lists))),y,color='blue')
plt.xlabel('구')
plt.ylabel('범죄 발생 수')
plt.savefig('data.png')
plt.show()
f.close()
```
