```python
import csv
import matplotlib.pyplot as plt
f = open('2010__ (1).csv')
data = csv.reader(f)
add = 0
y=[]
dummy1 = []
dummy2 = 0
for i in range(0,25) :
    y.append(0)
lists = ['', '', '', '', '', '', '', '', '', '', '', '', '', '', '','', '', '', '', '', '', '', '', '', '']
for row in data :
    for k in range(0,len(lists)) :
        if lists[k] in row[0] :
            dummy1 = y[k+1:]
            y = y[:k+1]
            dummy2 = y[k]
            y.pop()
            for i in list(range(1,len(row))) :
                d = row[i].replace(',','')
                d = d.replace('-','0')
                dummy2 += int(d)
            y.append(dummy2)
            y = y + dummy1
            dummy2 = 0
plt.figure(figsize = (25,30), dpi = 300)
plt.rc('font',family='Malgun Gothic')
plt.title('  ')
plt.xticks(range(25),lists)
plt.bar(list(range(0,len(lists))),y,color='blue')
plt.xlabel('')
plt.ylabel('  ')
plt.savefig('data.png')
plt.show()
f.close()
```
